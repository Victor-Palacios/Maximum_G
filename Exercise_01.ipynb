# Table C Answer; A function that can be dragged by both row and columns in excel would be most similar to a python function as demonstrated below.

```python
df1["Revenue before discount.1"] = df1["Sales (Quantity).2"]*df1["Sales Price.1"]

def discount_function(discount_scenario):
    df1[discount_scenario][1:] = df1["Sales (Quantity).2"][1:]*(
                                    df1["Sales Price.1"][1:]*(
                                        1-df1[discount_scenario][0]))
    
    return df1.loc[:,"Sales (Quantity).2":"Discount Scenario 3"]
```

### Demonstration of Function

```python
discount_function("Discount Scenario 1")
discount_function("Discount Scenario 2")
discount_function("Discount Scenario 3")
```

# Table D Answer: Another function with only one input (like a drop down menu) would work here as well.
